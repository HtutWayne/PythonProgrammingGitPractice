```python
#Making Bad Students Game

import pygame
import random
import time

# Initialize Pygame
pygame.init()

# Set up some constants
WIDTH, HEIGHT = 800, 600
GREEN = (0, 255, 0)
RED = (255, 0, 0)

class Dot:
    def __init__(self, color, x, y):
        self.color = color
        self.x = x
        self.y = y
        self.radius = 20  # Increased radius
        self.vx = random.choice([-2, 2])  # Decreased velocity
        self.vy = random.choice([-2, 2])  # Decreased velocity
        self.collisions = 0

    def move(self):
        self.x += self.vx
        self.y += self.vy

        if not 0 < self.x < WIDTH:
            self.vx *= -1
        if not 0 < self.y < HEIGHT:
            self.vy *= -1

    def draw(self, screen):
        pygame.draw.circle(screen, self.color, (self.x, self.y), self.radius)

    def check_collision(self, other):
        return ((self.x-other.x)**2 + (self.y-other.y)**2) ** 0.5 <= self.radius + other.radius

# Create dots
dots = [Dot(RED, random.randrange(WIDTH), random.randrange(HEIGHT))]
dots.extend(Dot(GREEN, random.randrange(WIDTH), random.randrange(HEIGHT)) for _ in range(11))

# Set up the display
screen = pygame.display.set_mode((WIDTH, HEIGHT))

# Set up the clock
clock = pygame.time.Clock()

# Game loop
start_ticks = pygame.time.get_ticks()
while True:
    # Check for events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
        elif event.type == pygame.MOUSEBUTTONDOWN:
            x, y = pygame.mouse.get_pos()
            for dot in dots:
                if dot.color == RED and ((x-dot.x)**2 + (y-dot.y)**2) ** 0.5 <= dot.radius:
                    dots.remove(dot)
                    break

    # Move and draw each dot
    screen.fill((0, 0, 0))
    for dot in dots:
        dot.move()
        dot.draw(screen)

    # Check for collisions
    for dot in dots:
        if dot.color == RED:
            for other in dots:
                if other.color == GREEN and dot.check_collision(other):
                    other.collisions += 1
                    if other.collisions == 3:
                        other.color = RED

    # Update the display
    pygame.display.flip()

    # Cap the frame rate
    clock.tick(30)  # Decreased frame rate

    # Check for end of game
    seconds = (pygame.time.get_ticks() - start_ticks) / 1000

    if seconds >= 300:  # 5 minutes
        reds = sum(1 for dot in dots if dot.color == RED)
        greens = sum(1 for dot in dots if dot.color == GREEN)
        print('Game Over!')
        if reds > greens:
            print('You lose!')
        else:
            print('You win!')
        pygame.quit() #Finished for timeframe
  
       


```
